## Практическое задание

__1.__ Исследовать на линейную зависимость:
$$f_{1}(x)=e^{x}, f_{2}(x)=1, f_{3}(x)=x+1, f_{4}(x)=x-e^{x}.$$

### Решение

Заметим, что f4(x)=f3(x)-f2(x)-f1(x), то есть вектор f4(x) — линейная комбинация векторов f1(x), f2(x) и f3(x), из чего можно сделать вывод, что f4(x), f3(x), f2(x), f1(x) линейно зависимы.

__2.__ Исследовать на линейную зависимость:
$$f_{1}(x)=2, f_{2}(x)=x, f_{3}(x)=x^{2}, f_{4}(x)=(x+1)^{2}.$$

### Решение
Возведенем в квадрат $$f_{4}(x)=(x+1)^{2} = x^{2} + 2x + 1.$$
Заметим, что f4(x)=f3(x)+2f2(x)+f1(x)-1, то есть вектор f4(x) — линейная комбинация векторов f1(x), f2(x) и f3(x), из чего можно сделать вывод, что f4(x), f3(x), f2(x), f1(x) линейно зависимы.


__3.__ Найти координаты вектора $x = (2, 3, 5)\in \mathbb{R}^{3}$ в базисе $b_{1}=(0, 0, 10)$, $b_{2}=(2, 0, 0)$, $b_{3}=(0, 1, 0)$.

### Решение

Стандартный базис линейного пространства R3 образуют векторы b1=(0,0,10), b2=(2,0,0), b3=(0,1,0). Тогда

x=(2,3,5)=(2,0,0)+(0,3,0)+(0,0,5)=2⋅(1,0,0)+3⋅(0,1,0)+5⋅(0,0,1)=2b2+3b3+0.5b1,
то есть координатами вектора x в стандартном базисе являются 0.5, 2, 3.

__4.__ Найти координаты вектора $3x^{2}-2x+2\in\mathbb{R}^{3}[x]$:

### Решение:

a) координатами вектора x в стандартном базисе являются 2, -2, 3;

б) координатами вектора x в стандартном базисе являются 3, -2, 0.
